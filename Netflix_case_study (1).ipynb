```python
# convert the data type from object to datetime64
df["date_added"]=pd.to_datetime(df['date_added'])
```

```python
# Data Cleaning (Optional Treatment) - Check for Missing values & Duplicates.
```

```python
# null counts
df.isnull().sum().sort_values(ascending=False)
```

```python
# Null values percentage calculation
round((df.isnull().sum()/len(df.index))*100,2).sort_values(ascending=False)
```

```python
# drop low percentage null values
df=df[~pd.isnull(df["rating"])]
df=df[~pd.isnull(df["date_added"])]
df=df[~pd.isnull(df["duration"])]
```

```python
# Replacing the null values for country,cast,director
df["country"].replace(np.nan,"No country",inplace=True)
df["cast"].replace(np.nan,"No cast",inplace=True)
df["director"].replace(np.nan,"No director",inplace=True)
```

```python
# checking for the null count again
df.isnull().sum().sort_values(ascending=False)
```

```python
# Non-Graphical Analysis
df.head()
```

```python
# Unnesting of director columns & Fetching top 5 Directors -
filtered_directors=pd.DataFrame()
```

```python
filtered_directors=df["director"].str.split(",",expand=True).stack()
```

```python
filtered_directors=filtered_directors.to_frame()
filtered_directors.columns=["Director"]
directors=filtered_directors.groupby(["Director"]).size().reset_index(name="Total content")
directors=directors[directors.Director!="No director"]
directors=directors.sort_values(["Total content"],ascending=False)
```

```python
directors.head()
```

```python
# Unnesting of cast columns & Fetching top 5 actors -
filtered_cast=pd.DataFrame()
filtered_cast=df["cast"].str.split(",",expand=True).stack()
filtered_cast=filtered_cast.to_frame()
filtered_cast.columns=["Actor"]
actor=filtered_cast.groupby(["Actor"]).size().reset_index(name="total content")
actor=actor[actor.Actor!="No cast"]
actor=actor.sort_values(["total content"],ascending=False)
actor.head()


```

```python
# Unnesting of country columns & Fetching top 5 actors -
filtered_country=pd.DataFrame()
filtered_country=df["country"].str.split(",",expand=True).stack()
filtered_country=filtered_country.to_frame()
filtered_country.columns=["Countries"]
Country=filtered_country.groupby(["Countries"]).size().reset_index(name="totcount")
Country=Country[Country.Countries!="No country"]
Country=Country.sort_values(["totcount"],ascending=False)
Country.head()

```

```python
# lineplot Approach-
plt.figure(figsize=(10,6))
df1=df[["type","release_year"]]
df2=df1.groupby(["type","release_year"]).size().reset_index(name="total_content")
df2=df2[df2["release_year"]>=1990]
sns.lineplot(data=df2,x="release_year",y="total_content",hue="type")
plt.show()
```

```python
# Insights:-
```

<h4> Line plot clearly shows that maximum Movies and TV shows are release after 2015 on netflix

```python
# Recent trend in movies vs Tv shows:
df1=df[["release_year","type"]]
df2=df1.groupby(["release_year","type"]).size().reset_index(name="total_content")
df2=df2[df2["release_year"]>=2015]
df2
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df2,x="release_year",y="total_content",hue="type")
plt.title("Recent trend in Movies vs TV show")
plt.show()
```

```python
# Insights:-
```

<h4> 1) Movies line plot got hump in between 2017 to 2018 which means the count was at its peak point.
</h4>
<h4> 2) For TV Shows the count was increased after 2015 but with lesser slope as compaired to movies.<\h4>
<h4> 3) But after 2020 both movies and TV shows experienced a sudden falldown.</h4>

```python
# Best time to lauch TV Show and Movies:
df["month"]=df["date_added"].dt.month
df["month"].value_counts().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x="month",order=df["month"].value_counts().index[:12])
plt.show()
```

```python
# Insights:-
```

<h4> From the above plot it is clear that maximum movies and Tv shows are added in the netflix in the month of July followed by December.</h4>

```python
# Gap between release year and date_added
df["dateadd"]=df["date_added"].dt.year
df["yeargap"]=df["dateadd"]-df["release_year"]
plt.figure(figsize=(10,6))
sns.countplot(data=df,y="yeargap",order=df["yeargap"].value_counts().index[:10])
plt.show()
```

```python
# Insights:
```

<h4> From the above countplot it is clear that a large number of movies and TV show are added to netflix with zero year gap between release year and added_year.

```python
# Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(x="type",y="yeargap",data=df)
plt.show()
```

<h4> From the above boxplot,median signifies that there are movies and TV show whose release year and year of being added to netflix is almost zero.

```python
# Correlation
df.corr()
```

```python
new_df=df.groupby("dateadd")["month"].value_counts().unstack().T
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.pcolor(new_df,cmap="coolwarm",edgecolor="White",linewidth=2)
plt.xticks(np.arange(0.5,len(new_df.columns),1),new_df.columns,fontsize=7)
plt.yticks(np.arange(0.5,len(new_df.index),1),new_df.index,fontsize=7)
plt.title("Month Vs Netflix content update",fontsize=12)
cbar=plt.colorbar()
cbar.ax.tick_params(labelsize=8)
cbar.ax.minorticks_on()
plt.show()
```

```python
# Insights:
```

<h4> 1)The above heatmap shows relationship between months and the content added to the Netflix.
2)From the above heat map it is clear that maximum movies and Tv shows are added to the Netflix in the july 2021 and December 2019.</h4>
